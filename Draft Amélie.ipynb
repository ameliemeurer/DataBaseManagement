We put in the RDDfile the data set rating, and we only keep the 2 first columns

```python
df = sc.textFile("data_small/ratings.csv").map(lambda line: line.split(",")).map(lambda line: (line[0],line[1]))
```

```python
head= df.first()
header = sc.parallelize([head])
df = df.subtract(header)
```

```python
df = df.map(lambda line : (int(line[0]), int(line[1])))
```

```python
df.collect()
```

```python
def cartesian_group(i): 
    groups = sc.parallelize(np.arange(1, i+1))
    cartesian2_df = groups.cartesian(df)
    cartesian2_df = cartesian2_df.map(lambda line : (line[0], (line[1])[0], (line[1])[1]))
    return(cartesian2_df)
```

```python
df_cart = cartesian_group(3)
```

```python
df_cart.collect()
```

```python
df_cart.count()
```

```python
data_init = df_cart.map(lambda x : (x, np.random.random())).map(lambda line : ((line[0])[0], (line[0])[1], (line[0])[2], line[1]))
```

```python
data_init.collect()
```
